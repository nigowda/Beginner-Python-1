# **Introduction to Data types**


- The type of data that we are delaing with eg: **Text**, **numbers**, **decimals** etc.
- Each data type has a specific attribute

## **Common Datatypes in Python**



### **Int (Integers)**



- A data type that deals with **positive** or **negative** whole numbers.
eg: 1, -4, 8, 0, 323434242...

- Python has no upper limit on size.


### **Float (Decimals)**


All real numbers that are not Integers.
eg: .3, 0.123, -2.4

#### **Question**: Is 9 and 9.0 same?
In real sense they are same, but in python they are **different**, 9.0 is a float and 9 is an integer.

'.' is not a decimal but '0.' is

### **Strings**



- Anything inside quote are considered string. We can use Double quotes(`" "`) or single quuotes (`' '`).
- We have to start and end with the same type of quote.

**Example**:
- `"Hello everyone"` - **valid**
- `'hello'` - **valid**
- `"hey everyone'` - **invalid**

### **Bool (boolean)**



- It has a `True` or `False` value.
- `True` is usualy 'high' and `False` is 'low'.
- We using boolean values when comparing.

### **None** type




- Has only one value, `None`.
- To represent nothing or an empty value, we use `None`.
