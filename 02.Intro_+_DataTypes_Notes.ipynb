### **How do you check data type of a given object?**

 #### ```type()```
 - type function takes in an object and returns the data type of the object.
 - It gives the result, but does not print it.

````python
print("Some random string") #prints"some random string"
````

````python
type("Some random string") #Runs, but does not print"
````

````python
type(45)
````

````python
type(67.8)
````

````python
type(True)
````

* IDE or a python shell might not print the result from `type()`.

* But, jupyter or colab notebook prints the result of `type()` by default.

````python
print(type("hello"))
````

* The result of type() function is passed to the print() function and the output is printed on the screen, in both IDE and colab notebook.

  * Digits (0 to 9)
  * Underscore (_)

## **Introduction to Operators**



- Operators are symbols of operation.
- Operands are values on which operation is happening
- Operation/Expression are combination of operands and operators
- Operation/Expression gives a result after execution.




image.png

### Common Operator types

Operators are usually of three types:

1. ***Arithmetic Operators*** - Used to perform mathematical operations. Example: **Plus** (+), **Minus** (-), etc
2. ***Comparison Operators*** - Used to perform comparisons like **greater than** (>), **less than**(<) or **equal to** (=)
3. ***Boolean Operators*** - Logical terms used to combine expressions. Example: **`and`**, **`or`**, **`not`**
